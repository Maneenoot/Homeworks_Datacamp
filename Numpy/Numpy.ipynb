20. เขียนโปรแกรมรับค่าจากผู้ใช้ให้ป้อนเลข 9 ตัว แต่ละตัวต้องไม่ซ้ำกันและมีค่าอยู่ในช่วง 10-100 นำเลข 9 ตัวนั้นมา
ทำเป็น matrix 3x3 และ transport และหาเลขอยู่ตำแหน่งที่ (2,2)

```python
p1=[]
for i in range(9):
    number=int(input('input number 10-100: '))
    i += 1
    p1.append(number)
print(list(p1))    
```

```python
p2=np.array(p1)
p3=p2.reshape(3,3)
p3
```

```python
ztranspose = p3.transpose()
ztranspose
```
